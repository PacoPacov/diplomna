```python
from youtube_transcript_api import YouTubeTranscriptApi
import json
import os


transcript_list = YouTubeTranscriptApi.list_transcripts("neVtNFPTsiQ", 'en')

auto_generated_trans = []

for trans in transcript_list:
    print(trans.language, trans.is_generated)
    if trans.is_generated:
        auto_generated_trans.append(trans.fetch())

if auto_generated_trans:
    output_path = os.path.join("/home/paco/Documents/diplomna/transcripts/", 'neVtNFPTsiQ_en_transcript.json')
    print(output_path)
    with open(output_path, 'w') as f:
        json.dump(auto_generated_trans[0], f, indent=4)
```

```python
import re

def check_transcipt(transcript, annotation_row):
    for row in transcript:
        transcript_row = re.sub(r'[^\w\s]','',row['text'])
        annotation_row = re.sub(r'[^\w\s]','',annotation_row)

        print(annotation_row.lower(), "----", transcript_row.lower())
        if annotation_row.lower() in transcript_row.lower():
            return row['start']
        else:
            return None
```

```python
df['transcript_index'] = annotation_df['claim'].apply(lambda x: check_transcipt(transcript, x))
```
