```python
import os

dir_path = "/home/paco/Documents/diplomna/dataset/annotated_transcripts"

result = []
for file_ in os.listdir(dir_path):
    full_file_path = os.path.join(dir_path, file_)
    temp_df = pd.read_csv(full_file_path)
    value_counts_dict = temp_df['fact-checked'].value_counts().to_dict()
    print(file_, '--->', value_counts_dict)
    
    value_counts_dict['file_name'] = file_
    result.append(value_counts_dict)
```

```python
info_df = pd.DataFrame(result)
```
